```python
truck_stops = pd.read_csv('truck_stops.csv', header=None, names=['Longitude', 'Latitude', 'Truck Stop Name',
                                                                'Truck Stop Info', 'Phone Number','Street Address',
                                                                'City', 'State/Province', 'Postal Code', 'Country'])
truck_stops = truck_stops.drop([0.000000])

truck_stops.head()
```

```python
# Dump Stations Dataset

dump_stations = pd.read_csv('Raw POI Factory Datasets/Dump_Stations.csv', encoding='ISO-8859-1', header=None)

# Removing columns with null values
dump_stations = dump_stations.drop(columns=[3, 5, 6, 7, 8])

# Adding column header names
dump_stations = dump_stations.rename(columns={0: 'Longitude', 1: 'Latitude', 
                                              2: 'Dump Site Name', 4: 'State/Province'})

print(dump_stations.shape)
dump_stations.head()
```

```python
dump_stations.info()
```

```python
dump_stations[dump_stations.isna().any(axis=1)]
```

```python
dump_stations = dump_stations.fillna('AB')

dump_stations.info()
```

```python
# Writing cleaned dataset to csv
dump_stations.to_csv(os.path.join(r'dump_stations.csv'),index=False)
```
